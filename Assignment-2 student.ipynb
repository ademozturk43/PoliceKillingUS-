## 2: Plot 12 most dangerous cities

```python
city_freq = 
city_freq
```

```python
#kill.head()
city_freq = kill["city"].value_counts().head(12)
city_freq 
```

```python
city_freq= pd.DataFrame(data=city_freq)
```

```python
# with seaborn barplot
fig, ax =plt.subplots(figsize=(8,5))
ax = 
ax.
ax.
```

```python
# with seaborn barplot
fig, ax =plt.subplots(figsize=(8,5))
ax = sns.barplot(x = city_freq.index, y = city_freq["city"])
ax.set_title("Most dangerous cities", color = "b", size = 16)
ax.tick_params(axis='x', labelrotation=45)
ax.set(ylabel=None);
```

```python
# with seaborn countplot
fig, ax =plt.subplots(figsize=(8,5))
ax = sns.countplot(x = "city", data = kill, order=kill.city.value_counts().iloc[:12].index)
ax.set_title("Most dangerous cities", color = "b", size = 16)
ax.tick_params(axis='x', labelrotation=45)
```
